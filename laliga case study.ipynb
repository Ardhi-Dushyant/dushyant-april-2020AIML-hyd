### Question4:Write a function with the name “Goal_diff_count”
### which should return all the teams with their Goal Differences


```python
# method1:
def Goal_diff_count(x,y):
    z = x-y
    print(z) 
Goal_diff_count(df['GoalsFor'],df['GoalsAgainst'])
```

```python
# method2:
Goal_diff_count=lambda x,y:abs(x-y)
print(Goal_diff_count(df['GoalsFor'],df['GoalsAgainst']))
```

```python
# method3:
def Goal_diff_count(GF,GA):
    Goal_diff_count = GF-GA
    return Goal_diff_count
```

```python
df['Goal_diff_count'] = Goal_diff_count(df['GoalsFor'],df['GoalsAgainst'])
print (df[['Team','Goal_diff_count']])
```
